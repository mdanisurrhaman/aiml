7 Percentage of sales across each region

```python
total_sales
```

```python
Region_sales["Percentage_sales"]=(Region_sales["Sales"]/total_sales) *100
```

```python
Region_sales
```

```python
plt.pie(
    x=Region_sales["Sales"], labels=Region_sales["Region"].value_counts().index,
    autopct="%.2F"
)
```

Q10.  How many of the Top 50 customers by Sales are also the from bottom 50 customers by Profit.

```python
Top_50Customer=df.groupby("Customer Name").agg({"Sales":"sum"}).sort_values(by="Sales", ascending=False).head(50)
```

```python
Top_50Customer
```

```python
df.groupby("Customer Name").agg({"Sales":"sum"}).sort_values(by="Sales")
```

```python
df.groupby("Customer Name").agg({"Profit":"sum"}).sort_values(by="Profit")
```

```python
bott_50Customer=df.groupby("Customer Name").agg({"Profit":"sum"}).sort_values(by="Profit").head(50)
```

```python
bott_50Customer
```

```python
Top_50Customer
```
