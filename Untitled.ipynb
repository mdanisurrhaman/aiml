```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('CancerDataset.csv')
df.drop("Unnamed: 32",  axis=1, inplace=True )
df.head()
```

```python
X=df.drop("diagnosis", axis=1)
y=df["diagnosis"]
```

```python
X.head()
```

```python
y=y.map(lambda x:1 if x=="M" else  0 )
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
X_tr, X_cv, y_tr, y_cv= train_test_split(
    X_train, y_train, test_size=0.32, random_state=42)
```

```python
df.shape
```

```python
X_tr.shape, X_cv.shape, X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model=KNeighborsClassifier()
```

```python
knn_model.fit(X_tr , y_tr)
```

```python
y_pred=knn_model.predict(X_tr)
```

```python
y_tr,  y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_tr , y_pred)
```

```python
y_pred_cv=knn_model.predict(X_cv)
accuracy_score(y_cv , y_pred_cv)
```

```python
neighbors=range(1,30)
train_accuracies=[]
cv_accuracies=[]
test_accuracies=[]
for k in neighbors:

    knn_model=KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_tr , y_tr)
    train_accuracies.append(accuracy_score(y_tr , knn_model.predict(X_tr) ))
    cv_accuracies.append(accuracy_score(y_cv , knn_model.predict(X_cv) ))
    test_accuracies.append(accuracy_score(y_test, knn_model.predict(X_test)))


print(train_accuracies)
print(cv_accuracies)
print(test_accuracies)
```

```python
cv_accuracies
```

```python
test_accuracies
```

```python
sns.lineplot(y=train_accuracies , x=neighbors , label= "train_accuracies")
sns.lineplot(y=cv_accuracies , x=neighbors , label="cv_accuracies" )
sns.lineplot(y=test_accuracies , x=neighbors , label="test_accuracies" )

plt.legend()
plt.show()
```
